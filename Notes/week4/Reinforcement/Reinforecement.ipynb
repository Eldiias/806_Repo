```python
def categorize_strings(df, col, required_categories=[]):
    #create a copy of df
    df=df.copy()
    
    #exploratory
    print(f'The number of unique values in the column {col} is {df[col].nunique()}')
    display(df[col].describe())
    
    #create a list of words
    low=[j for i in df[col].str.lower().dropna().values for j in i.split()]
    print(len(low), 'values')
    print(len(set(low)), 'unique values')
    
    #top 30 categories
    categories=set(pd.Series([i.strip('ing ') for i in low]).value_counts().iloc[:30].index)
    
    #top 30 categories clean
    categories=categories-set(['','&','/','a','by','the','to','o','of','with','for','overboard','shark','from','free','fell','swimm'])
    categories=categories.union(set(required_categories))
    
    df[col].fillna('missing value', inplace=True)
    
    # a column with new categories
    for i in list(categories):
        df.loc[df[col].str.lower().str.contains(i), f'New_{col}']=i
    
    df[f'New_{col}'].fillna('other',inplace=True)
    
    display(df[f'New_{col}'].value_counts())
    
    return df
```

```python
df.drop(['New_Activity'], axis=1, inplace=True)
```

```python
def sample_sd(x):
    '''
    Input: A Pandas series of values
    Output: the standard deviation divided by the number of elements in the series
        
    Example:
    Input: pd.Series([1,2,3,4])
    Output: 0.3726779962
    '''
    
    # Your code here:
    return np.std(x)/(x.count()-1)
```
